```python
import cv2
import numpy as np
from PIL import Image

def add_snow(input_image, output_image, snow_density=1000):
    # Load the image
    image = cv2.imread(input_image)
    h, w, _ = image.shape

    # Create a blank snow layer
    snow_layer = np.zeros((h, w, 3), dtype=np.uint8)

    # Generate random snowflakes
    for _ in range(snow_density):
        x, y = np.random.randint(0, w), np.random.randint(0, h)
        size = np.random.randint(1, 4)  # Snowflake size
        cv2.circle(snow_layer, (x, y), size, (255, 255, 255), -1)

    # Blend snow layer with the image
    alpha = 0.3  # Adjust intensity
    snowy_image = cv2.addWeighted(image, 1, snow_layer, alpha, 0)

    # Save and display
    cv2.imwrite(output_image, snowy_image)
    Image.open(output_image).show()

# Run snow effect on uploaded image
add_snow("input.jpg", "opencv_snow.jpg")
print("✅ OpenCV snow effect saved as 'opencv_snow.jpg'")
```

```python
from diffusers import StableDiffusionImg2ImgPipeline
import torch

# Load the Stable Diffusion img2img model
pipe = StableDiffusionImg2ImgPipeline.from_pretrained("runwayml/stable-diffusion-v1-5").to("cuda")

def generate_realistic_snow(input_image, output_image, strength=0.5):
    # Load the image
    image = Image.open(input_image).convert("RGB")

    # Run Stable Diffusion for realistic snow
    prompt = "A beautiful snowy winter scene, soft falling snow, white landscape, high realism"
    snowy_image = pipe(prompt=prompt, image=image, strength=strength, guidance_scale=7.5).images[0]

    # Save and display
    snowy_image.save(output_image)
    snowy_image.show()

# Apply Stable Diffusion to refine OpenCV snow
generate_realistic_snow("opencv_snow.jpg", "realistic_snow.jpg")
print("✅ Realistic snow image saved as 'realistic_snow.jpg'")
```
